```python
import json
def read_jsonl(file_path):
    data = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            data.append(json.loads(line))
    return data
##please edit your path to dev_tables.jsonl here
file_path = 'dev_tables.jsonl'
jsonl_data = read_jsonl(file_path)
```

```python
import pandas as pd
def process_one_table(data):
    # processed headers that will be used
    processed_table_headers = data['processed_tableHeaders']
    table_data = data['tableData']
    # Convert the table data to a list of lists with only the 'text' values
    formatted_table_data = [[cell['text'] for cell in row] for row in table_data]
    # Create a pandas DataFrame using the processed_tableHeaders and formatted_table_data
    df = pd.DataFrame(formatted_table_data, columns=processed_table_headers)
    return df
# df = df.drop('name', axis=1)

process_one_table(jsonl_data[5])
```

```python
tables= []
for data in jsonl_data:
    tables.append(process_one_table(data))
```

```python
import os
tables_copy = tables.copy()
directory = "tables_csv"
if not os.path.exists(directory):
    os.makedirs(directory)
for i, table in enumerate(tables):
    table.to_csv(f"{directory}/table_{i}.csv", index=False, na_rep='')

# Load tables from CSV files
tables2 = []
for i in range(len(tables)):
    table = pd.read_csv(f"{directory}/table_{i}.csv", keep_default_na=False)
    tables2.append(table)

```
